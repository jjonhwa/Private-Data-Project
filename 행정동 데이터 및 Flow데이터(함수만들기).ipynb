### 함수

```python
def open_file(file_name) :
    name = pd.read_csv('../data/01_유동인구데이터(SK텔레콤)/%s' % file_name, sep = '|')
    return name
```

```python
def STD_YM_drop(file) :
    file.drop(['STD_YM'], axis = 1, inplace = True)
    file.rename(columns = {'STD_YMD' : '기준년월일'}, inplace = True)
    return file
```

```python
def make_new_column(file) :
    si_name = []
    gu_name = []
    
    for i in range(len(file)) :
        if str(file['HDONG_CD'][i])[:2] == '11' :
            si_name.append('서울광역시')
        else :
            si_name.append('대구광역시')
            
    for i in range(len(file['HDONG_CD'])) :
        if str(file['HDONG_CD'][i])[:5] == '27260' :
            gu_name.append('수성구')
        elif str(file['HDONG_CD'][i])[:5] == '27110':
            gu_name.append('중구')
        elif str(file['HDONG_CD'][i])[:5] == '11350' :
            gu_name.append('노원구')
        else :
            gu_name.append('중구')
            
    region = []
    
    for i in range(len(file)) :
        region.append(si_name[i] + ' ' + gu_name[i] + ' '+ file['HDONG_NM'][i])
        
    file['지역명'] = region
    file.index = file['지역명']
    file.drop(['지역명'], axis = 1, inplace = True)
    file.reset_index(drop = False, inplace  = True)
    
    return file
```

```python
def drop_columns(file) :
    file.drop(['HDONG_CD', 'HDONG_NM'], axis = 1, inplace = True)
    return file
```

```python
def make_columns_name(file) :
    age = ['지역명', '기준년월일']

    for i in range(len(file.columns)) :
        if file.columns[i][-1] == 'U' and file.columns[i][0] == 'M':
            age.append('남자 70세 이상')
        
        elif file.columns[i][0] == 'M' :
            age.append('남자' + ' ' + str(file.columns[i][-4:-2]) + '세~' + str(file.columns[i][-2:]) + '세')
        
        elif file.columns[i][-1] == 'U' and file.columns[i][0] == 'W' :
            age.append('여자 70세 이상')
        
        elif file.columns[i][0] == 'W' :
            age.append('여자' + ' ' + str(file.columns[i][-4:-2]) + '세~' + str(file.columns[i][-2:]) + '세')
            
    file.columns = age
    
    return file
```

```python
def save_file(file, name) :
    file.to_excel('../data/%s' % name, index = False)
```

### 파일 만들기

```python
import glob
```

```python
path = '../data/01_유동인구데이터(SK텔레콤)'

file_list = glob.glob(path + '\\*.csv')
file_list
```

```python
print(file_list[0][26:46])
print(len(file_list[0][26:46]))
```

```python
naming = []

for i in range(len(file_list)) :
    if len(file_list[i][26:46]) == 20:
        naming.append(file_list[i][26:46])
```

```python
naming = naming[:8]
naming
```

```python
print(naming[0][-6:-2])
print(naming[0][-2:])
```

```python
def file_parsing(file_list) :
    
    name_list = pd.DataFrame(file_list)
    name_list = name_list + '.csv'
    name_list = name_list.iloc[:, 0]
    
    for i in range(len(name_list)) :
        flow = open_file(str(name_list[i]))
        flow = STD_YM_drop(flow)
        flow = make_new_column(flow)
        flow = drop_columns(flow)
        flow = make_columns_name(flow)
        
        save_file(flow, '유동인구_나이{}년{}월.xlsx'.format(file_list[i][-6:-2], file_list[i][-2:]))
```

```python
file_parsing(naming)
```
